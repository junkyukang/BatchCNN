label 파트는 학습시 사용될 label 과 segment 구분을위한 데이터를 전처리하는 단계

아래 작업들은 발화 당사자의 감정예측을 분류하고, 결측치, 중복되는 감정을 제외하는 과정임

```python
import os
files_F=[]
files_M=[]

files = os.listdir('./KEMDy19/annotation')
for i in range(len(files)):
    if files[i][10]=='F':
        files_F.append(files[i])
    else:
        files_M.append(files[i])
```

```python
data_F_res = pd.concat([pd.read_csv(f"./KEMDy19/annotation/{j}") for j in files_F], ignore_index=True)
data_M_res = pd.concat([pd.read_csv(f"./KEMDy19/annotation/{j}") for j in files_M], ignore_index=True)

data_F_res
```

```python
data_F_res= data_F_res[(data_F_res["Total Evaluation"]=='neutral') | (data_F_res["Total Evaluation"]=='angry') | (data_F_res["Total Evaluation"]=='happy') | (data_F_res["Total Evaluation"]=='surprise') | (data_F_res["Total Evaluation"]=='sad') | (data_F_res["Total Evaluation"]=='fear') | (data_F_res["Total Evaluation"]=='disgust') ]
data_M_res= data_M_res[(data_M_res["Total Evaluation"]=='neutral') | (data_M_res["Total Evaluation"]=='angry') | (data_M_res["Total Evaluation"]=='happy') | (data_M_res["Total Evaluation"]=='surprise') | (data_M_res["Total Evaluation"]=='sad') | (data_M_res["Total Evaluation"]=='fear') | (data_M_res["Total Evaluation"]=='disgust') ]
data_F_res=data_F_res.loc[:,['Segment ID','Total Evaluation']]
data_M_res=data_M_res.loc[:,['Segment ID','Total Evaluation']]
```

```python
data_F_res=data_F_res[data_F_res['Segment ID'].str[-4]=='F']
data_M_res=data_M_res[data_M_res['Segment ID'].str[-4]=='M']
```

```python
import os
import time
import re
from tqdm.auto import tqdm as tqdm_auto
import numpy as np
import pandas as pd

import audiofile
import opensmile
```

```python
dir_path = "./KEMDy19/wav/"
file_path = []

for (root, directories, files) in os.walk(dir_path):
    for file in files:
        if '.wav' in file:
            file_ = os.path.join(root, file)
            file_path.append((file_,file))
```

```python
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
)
```

```python
z=[0 for i in range(len(file_path))]
for i in tqdm_auto(range(len(file_path))):
    y = smile.process_file(file_path[i][0])    
    y.insert(0,'Segment ID', file_path[i][1][0:-4]) #해당 음성의 session 정보 / .wav는 session 이름에서 삭제 
    z[i]=y
acoustic_data= pd.concat(z,ignore_index=True)
```

```python
acoustic_data
```

```python
merge_data_1= pd.merge(data_F_res,acoustic_data)
merge_data_2= pd.merge(data_M_res,acoustic_data)
```

```python
merge_data_1=merge_data_1.dropna()
merge_data_2=merge_data_2.dropna()
```

```python
merge_data= pd.concat([merge_data_1, merge_data_2], ignore_index=True)
```

```python
audio_data=[0 for i in range(len(merge_data))]

for i in tqdm_auto(range(len(merge_data))):
    audio_data[i]=np.array(merge_data.iloc[i,2:])
merge_data['audio_data']=audio_data
```

```python
merge_data=merge_data[['Segment ID','Total Evaluation','audio_data']]
```

```python
merge_data
```

<h1>Text</h1>
발화문의 데이터를 추출하는 과정

```python
dir_path = "./KEMDy19/wav/"
file_path = []

for (root, directories, files) in os.walk(dir_path):
    for file in files:
        if '.txt' in file:
            file_ = os.path.join(root, file)
            file_path.append((file_,file))
```

```python
session=[0 for i in range(len(file_path))]
data=[0 for i in range(len(file_path))]
for i in tqdm_auto(range(len(file_path))):
    f = open(file_path[i][0], 'r')   
    line = f.readline()
    data[i]=line
    session[i]=file_path[i][1][:-4]
    f.close()
    
session
```

```python
for i in range(len(data)):
    data[i] = re.sub(r"\n|o/|c/|n/|N/|u/|l/|b/|/|\*|\+|\/|\(|\)|", "", data[i])
```

```python
Text_data=pd.DataFrame({'Segment ID': session ,'text_data': data})
Text_data
```

```python
final_data= pd.merge(merge_data,Text_data)
```

```python
final_data
```

```python
def change(x):
    if x=='fear': return 0                     
    elif x== 'disgust': return 1     
    elif x== 'sad': return 2  
    elif x== 'happy': return 3      
    elif x== 'surprise': return 4     
    elif x== 'angry': return 5                           
    elif x== 'neutral': return 6   
```

```python
final_data['label']= final_data['Total Evaluation'].apply(change)
```

```python
train_data=final_data[~((final_data['Segment ID'].str[4:6]=='17')| (final_data['Segment ID'].str[4:6]=='18') | (final_data['Segment ID'].str[4:6]=='19') | (final_data['Segment ID'].str[4:6]=='20'))]
test_data=final_data[(final_data['Segment ID'].str[4:6]=='17')| (final_data['Segment ID'].str[4:6]=='18') | (final_data['Segment ID'].str[4:6]=='19') | (final_data['Segment ID'].str[4:6]=='20')]
```

```python
train_data.dropna()
test_data.dropna()
test_data=test_data.reset_index(drop=True)
train_data=train_data.reset_index(drop=True)
```

```python
test_data
```

```python
#ecg, eda 데이터까지 온전한 데이터 sampling을 위해 ecg, eda 데이터도 merge 해줌
#fft데이터는 첨부되어 있음
ecg= pd.read_pickle("fft_ecg_data.pickle") 
eda= pd.read_pickle("fft_eda_data.pickle")
```

```python
train_data=pd.merge(train_data,ecg)
train_data=pd.merge(train_data,eda)
train_data
```

```python
test_data=pd.merge(test_data,ecg)
test_data=pd.merge(test_data,eda)
test_data
```
